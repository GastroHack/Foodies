```python
# the database that comes from Dishtracker:
# it comes in json format

import pandas as pd
import numpy as np

data = pd.read_json("./database.json")
data
```

```python
# predict how many pancakes are needed for 1 day

# average amount of pancakes consumed in a day

# first, select only pancake data
pancakes = data[data.item=='pancakes']
pancakes
```

```python
import numpy as np

average_demand = int(np.round(np.mean(pancakes['consumed today'])))
average_demand

# this data goes into the dynamically updated database to the column "consumed on average"
```

```python
# let's make a plot of how we improved our production-waste discrepancy
# columns: produced | consumed | wasted
# 28 days, n 1&2 are sat and sun

prod = [40,80,80,80,60,40,40,
       60,80,80,70,60,50,50,
       65,80,75,60,50,55,55,
       65,85,80,50,52,53,57]
cons = [33,60,60,40,40,35,30,
       10,65,55,55,50,40,34,
       36,55,60,50,45,35,45,
       61,82,78,48,45,50,55]
```

```python
prod = np.array(prod)
cons = np.array(cons)
wast = prod - cons
```

```python
wast
```

```python
array = np.array([prod,cons,wast]).T
```

```python
import pandas as pd 
export = pd.DataFrame(array,columns = ['produced','consumed','wasted'])
export
```

```python
export.to_json("./export_for_plot.json")
```
